```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python


df = pd.read_csv("Fitness_Health_Tracking_Dataset.csv")

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.describe()
```

```python
print("Number of missing values in each column:")
print(df.isnull().sum())
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce", dayfirst=True)
```

```python
df.numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
df.numeric_columns
```

```python
df.groupby("User_ID")[[
    "Hours_Slept",
    "Water_Intake (Liters)",
    "Active_Minutes",
    "Heart_Rate (bpm)"
]].apply(lambda x: x.isnull().sum())
```

```python
numeric_cols = [
    "Hours_Slept",
    "Water_Intake (Liters)",
    "Active_Minutes",
    "Heart_Rate (bpm)"
]
```

```python
df[numeric_cols] = df.groupby("User_ID")[numeric_cols].transform(
    lambda x: x.interpolate(method="linear")
)
```

```python
df[numeric_cols] = df.groupby("User_ID")[numeric_cols].transform(
    lambda x: x.ffill().bfill()
)
```

```python
df["Workout_Type"] = df["Workout_Type"].fillna("No Workout")
```

```python
df["Date"] = df["Date"].fillna(method="ffill")+pd.to_timedelta(1, unit="D")
```
